```python
import pandas as pd
import random
from faker import Faker
import numpy as np
fake = Faker()
random.seed(42)
num_rows = 10000
id=[random.randint(1,5) for _ in range(num_rows)]
B_levels = [random.randint(15,100) for _ in range(num_rows)]
range_values = []
for battery in B_levels:
  correlated_range=20+((battery -15)/80)*400
  range_with_noise = correlated_range + random.uniform(-2,2)
  range_values.append(round(range_with_noise,2))
latitudes=[round(random.uniform(10.00,20.00),4) for _ in range(num_rows)]
longitudes= [round(random.uniform(76.00,80.00),4) for _ in range(num_rows)]
speed= [random.randint(30,200)for _ in range(num_rows)]

vehicle_status=[random.choice([0,1])for _ in range(num_rows)]
pincode=[fake.postcode() for _ in range(num_rows)]

mc=[random.randint(400,2500)for _ in range(num_rows)]
cc=[random.randint(10,100)for _ in range(num_rows)]

df=pd.DataFrame({
    'id':id,

    'Range': range_values,
    'Battery Level (%)':B_levels,
    'Latitude (N)':latitudes,
    'Longitude (E)':longitudes,

    'Speed':speed,
    'Vehicle Status':vehicle_status,
    'Pincode':pincode,
    'Maintenance Cost':mc,
    'Charge Cost':cc,

})

print(df.head(10))
```

```python
from datetime import date
def generate_random_dates(num_dates):
  fake=Faker()
  Dates=[]
  start_date=date(2024,1,1)
  end_date=date(2024,10,31)
  for _ in range(num_dates):
    random_date=fake.date_between(start_date=start_date,end_date=end_date)
    formatted_date=random_date.strftime("%d-%m-%y")
    Dates.append(formatted_date)
  return Dates
num_dates = 10000
random_dates = generate_random_dates(num_dates)
print(type(random_dates))
for date in random_dates:
  print(date)
df['Dates']=random_dates
df['Day']=df['Dates'].str[0:2]
df['Month']=df['Dates'].str[3:5]
df['Year']=df['Dates'].str[6:]
df.head(20)
df.to_csv('data1.csv')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv("chandana.csv")

# Convert dates into a datetime format and extract numerical representations
data['Dates'] = pd.to_datetime(data['Dates'])
data['Day'] = data['Dates'].dt.day
data['Month'] = data['Dates'].dt.month
data['Year'] = data['Dates'].dt.year

# Select relevant columns for feature set (X) and target variable (y)
features = ['Battery Level (%)', 'Speed', 'Acceleration 0-100 km/h', 
            'Top speed', 'Gross Vehicle Weight (GVWR)', 'Latitude (N)', 
            'Longitude (E)', 'Vehicle Status', 'Maintenance Cost']
X = data[features]
y = data['Range']

# Apply noise to the target 'Range' variable
dff = data.copy()
dff['Range'] = dff['Range'].apply(lambda x: x * 1.20 if x < 220 else x * 0.75)  # Introduce noise for diversity

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, dff['Range'], test_size=0.2, random_state=42)

# Standardize the features (optional but can help in some cases)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Mean Squared Error (Noisy Data): {mse}")
print(f"Root Mean Squared Error (Noisy Data): {rmse}")
print(f"Mean Absolute Error (Noisy Data): {mae}")
print(f"R² Score (Noisy Data): {r2}")

# Print coefficients and intercept to understand the model better
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# Check the predicted range for a few samples to visualize the noise's effect
print("Sample predictions with noise:")
print(f"Predicted for 1% battery: {model.predict(scaler.transform([[1, 50, 3.5, 200, 3000, 50, 70, 1, 200]]))}")
print(f"Predicted for 95% battery: {model.predict(scaler.transform([[95, 50, 3.5, 200, 3000, 50, 70, 1, 200]]))}")

# Optional: Clip the predictions to a reasonable range if needed
y_pred = np.clip(y_pred, 0, 500)  # Adjust to max realistic range, e.g., 500 km
```
